## 最適経路探索系

#### Treeデータ型の使い方

```python
from datatype.tree import TreeNode, Tree

# make level0
root = TreeNode()

# add level1
root.addChild(TreeNode()).addChild(TreeNode()).addChild(TreeNode())

# add level2
root[0].addChild(TreeNode())
root[1].addChild(TreeNode()).addChild(TreeNode())
root[2].addChild(TreeNode()).addChild(TreeNode())

# add level3
root[0][0].addChild(TreeNode()).addChild(TreeNode())
root[1][0].addChild(TreeNode())
root[2][1].addChild(TreeNode())
```

```python
tree = Tree(root)
```
