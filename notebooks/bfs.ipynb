```python
from datatype.linear import queue, stack
from datatype.tree import TreeNode
```

```python
def bfs(node: TreeNode):
    q = queue()
    q.enque(node)
    i = 0
    
    while not q.isEmpty():
        v = q.deque()
        v.value = i

        for child in v.children:
            if child.value is not None:
                continue
            else:
                q.enque(child)
                
        i += 1

    return node

def dfs(node: TreeNode):
    s = stack()
    s.push(node)
    i = 0

    while not s.isEmpty():
        v = s.pop()
        v.value = i

        for child in v.children:
            if child.value is not None:
                continue
            else:
                s.push(child)

        i += 1

    return node
```
